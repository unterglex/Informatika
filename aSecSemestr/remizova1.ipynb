```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

data = {
    'area': [28, 42, 45, 52, 56, 68, 70, 75, 90],
    'price': [3.10, 3.80, 3.90, 4.40, 4.50, 5.90, 5.60, 6.40, 7.30]
}
df = pd.DataFrame(data)
plt.scatter(df['area'], df['price'], color='red')
plt.xlabel('площадь (кв.м.)')
plt.ylabel('стоимость (млн. руб)')
plt.title('Зависимость цены от площади')
plt.show()
reg = linear_model.LinearRegression()
X_train = df[['area']].values
y_train = df['price'].values
reg.fit(X_train, y_train)
predicted_price_38 = reg.predict([[38]])
print(f"Предсказанная цена для квартиры 38 м²: {predicted_price_38[0]:.2f} млн. руб")
predicted_price_200 = reg.predict([[200]])
print(f"Предсказанная цена для квартиры 200 м²: {predicted_price_200[0]:.2f} млн. руб")
coef_a = reg.coef_[0]
coef_b = reg.intercept_
print(f"Коэффициент a: {coef_a:.4f}")
print(f"Коэффициент b: {coef_b:.4f}")
print(f"Уравнение модели: price = {coef_a:.4f} * area + {coef_b:.4f}")
plt.scatter(df['area'], df['price'], color='red', label='Данные')
plt.plot(df['area'], reg.predict(X_train), color='blue', label='Линия регрессии')
plt.xlabel('площадь (кв.м.)')
plt.ylabel('стоимость (млн. руб)')
plt.title('Линейная регрессия')
plt.legend()
plt.show()
new_data = {
    'area': [25, 37, 40, 58, 61, 67, 72, 85, 96, 105, 118]
}
pred = pd.DataFrame(new_data)
p = reg.predict(pred[['area']].values)
pred['predicted_price'] = p
print(pred)
pred.to_excel('predicted_prices.xlsx', index=False)
```
