```python
import matplotlib.pyplot as plt
import numpy as np

a = float(input("Введите коэффициент a: "))
b = float(input("Введите коэффициент b: "))
c = float(input("Введите коэффициент c: "))
start = float(input("Введите начальное значение x: "))
end = float(input("Введите конечное значение x: "))
num_points = int(input("Введите количество точек: "))

x = np.linspace(start, end, num_points)

y = a + b * x**(3/2) + c * x**3

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'y = {a} + {b}x^(3/2) + {c}x³', color='blue')
plt.title('График функции y = a + b·x^(3/2) + c·x³')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = []
y = []
with open('data13.txt', 'r') as file:
    for line in file:
        parts = line.split()
        x.append(float(parts[0]))
        y.append(float(parts[1]))

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.grid(True)

plt.show()
```
