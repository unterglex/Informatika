```python
import random
from math import sqrt
import matplotlib.pyplot as plt

p = 0.8
xMin1, yMin1 = 0, 0
xMax1, yMax1 = 50, 50
xMin2, yMin2 = 25, 25
xMax2, yMax2 = 75, 75
k = 3
pointsCount1 = 50
pointsCount2 = 50
x = []
y = []
for i in range(pointsCount1):
    x.append([random.uniform(xMin1, xMax1), random.uniform(yMin1, yMax1)])
    y.append(1)
for i in range(pointsCount2):
    x.append([random.uniform(xMin2, xMax2), random.uniform(yMin2, yMax2)])
    y.append(2)


def train_test_split(x, y, p):
    x_train = []
    y_train = []
    x_test = []
    y_test = []
    for i in range(round((pointsCount1 + pointsCount2) * p)):
        x_train.append(x[i])
        y_train.append(y[i])
    for j in range(round((pointsCount1 + pointsCount2) * (1 - p))):
        x_test.append(x[j])
        y_test.append(y[j])
    return x_train, x_test, y_train, y_test


x_train, x_test, y_train, y_test = train_test_split(x, y, p)


def fit(x_train, y_train, x_test, k):
    y_pred = []
    for test_point in x_test:
        distances = []
        for i in range(len(x_train)):
            dx = test_point[0] - x_train[i][0]
            dy = test_point[1] - x_train[i][1]
            dist = sqrt(dx ** 2 + dy ** 2)
            distances.append((dist, y_train[i]))
        distances.sort(key=lambda x: x[0])
        k_nearest = distances[:k]
        votes = {}
        for label in k_nearest:
            votes[label[1]] = votes.get(label[1], 0) + 1
        max_votes = max(votes.values())
        candidates = [cls for cls, cnt in votes.items() if cnt == max_votes]
        y_pred.append(random.choice(candidates))
    return y_pred


y_pred = fit(x_train, y_train, x_test, k)

accuracy = sum(1 for i in range(len(y_test)) if y_pred[i] == y_test[i]) / len(y_test)
print(f"Точность модели: {accuracy:.2f}")

plt.figure(figsize=(10, 6))
train_class1 = [p for p, lbl in zip(x_train, y_train) if lbl == 1]
train_class2 = [p for p, lbl in zip(x_train, y_train) if lbl == 2]
plt.scatter([p[0] for p in train_class1], [p[1] for p in train_class1],
            c='blue', marker='o', s=100, label='x_train для 1ого класса')
plt.scatter([p[0] for p in train_class2], [p[1] for p in train_class2],
            c='blue', marker='x', s=100, label='x_train для 2ого класса')
test_colors = []
test_markers = []
for true, pred in zip(y_test, y_pred):
    color = 'green' if true == pred else 'red'
    marker = 'o' if true == 1 else 'x'
    test_colors.append(color)
    test_markers.append(marker)

for i in range(len(x_test)):
    plt.scatter(x_test[i][0], x_test[i][1],
                c=test_colors[i],
                marker=test_markers[i],
                s=100,
                edgecolor='black')

plt.title(f'Визуализация классификации (Точность: {accuracy:.2f})')
plt.xlabel('Координата X')
plt.ylabel('Координата Y')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
from math import sqrt

# === 1. Генерация данных ===
# Параметры для первого класса (зеленые круги)
x_min1, y_min1 = 0, 0
x_max1, y_max1 = 50, 50

# Параметры для второго класса (синие крестики)
x_min2, y_min2 = 25, 25
x_max2, y_max2 = 75, 75

# Количество точек для каждого класса
points_count1 = 50
points_count2 = 50

# Создаем списки для координат и меток классов
coordinates = []  # Все точки
labels = []       # Соответствующие метки (0 или 1)

# Генерируем точки для первого класса (метка 0)
for _ in range(points_count1):
    x = random.uniform(x_min1, x_max1)
    y = random.uniform(y_min1, y_max1)
    coordinates.append([x, y])
    labels.append(0)  # 0 - первый класс

# Генерируем точки для второго класса (метка 1)
for _ in range(points_count2):
    x = random.uniform(x_min2, x_max2)
    y = random.uniform(y_min2, y_max2)
    coordinates.append([x, y])
    labels.append(1)  # 1 - второй класс

# === 2. Разделение на обучающую и тестовую выборки ===
def split_data(data, labels, train_percent):
    total = len(data)
    train_size = int(total * train_percent)
    
    # Обучающая выборка
    train_data = data[:train_size]
    train_labels = labels[:train_size]
    
    # Тестовая выборка
    test_data = data[train_size:]
    test_labels = labels[train_size:]
    
    return train_data, test_data, train_labels, test_labels

# Разделяем данные (80% обучение, 20% тест)
x_train, x_test, y_train, y_test = split_data(coordinates, labels, 0.8)

# === 3. Алгоритм k-ближайших соседей ===
def knn_predict(train_data, train_labels, test_data, k=3):
    predictions = []
    
    for test_point in test_data:
        # Рассчитываем расстояния до всех обучающих точек
        distances = []
        for i in range(len(train_data)):
            # Евклидово расстояние
            dx = test_point[0] - train_data[i][0]
            dy = test_point[1] - train_data[i][1]
            dist = sqrt(dx**2 + dy**2)
            distances.append( (dist, train_labels[i]) )
        
        # Сортируем по расстоянию и берем k ближайших
        distances.sort(key=lambda x: x[0])
        nearest = distances[:k]
        
        # Считаем голоса
        votes = {0: 0, 1: 0}
        for vote in nearest:
            votes[vote[1]] += 1
        
        # Определяем класс (случайно при равенстве голосов)
        if votes[0] > votes[1]:
            predictions.append(0)
        elif votes[1] > votes[0]:
            predictions.append(1)
        else:
            predictions.append(random.choice([0, 1]))
    
    return predictions

# Делаем предсказания
y_pred = knn_predict(x_train, y_train, x_test, k=3)

# === 4. Оценка точности ===
correct = 0
for true, pred in zip(y_test, y_pred):
    if true == pred:
        correct += 1

accuracy = correct / len(y_test)
print(f"Точность модели: {accuracy:.2f}")

# === 5. Визуализация ===
plt.figure(figsize=(10, 6))

# Отображаем обучающие данные
# Первый класс (0) - синие круги
train_0 = [p for p, lbl in zip(x_train, y_train) if lbl == 0]
plt.scatter([p[0] for p in train_0], 
            [p[1] for p in train_0],
            c='blue', marker='o', s=80, label='Обучение класс 0')

# Второй класс (1) - синие крестики
train_1 = [p for p, lbl in zip(x_train, y_train) if lbl == 1]
plt.scatter([p[0] for p in train_1], 
            [p[1] for p in train_1],
            c='blue', marker='x', s=80, label='Обучение класс 1')

# Отображаем тестовые данные
for i in range(len(x_test)):
    true_label = y_test[i]
    pred_label = y_pred[i]
    color = 'green' if y_test[i] == y_pred[i] else 'red'
    marker = 'o' if y_pred[i] == 0 else 'x'
    plt.scatter(x_test[i][0], x_test[i][1],
                c=color, marker=marker, s=80)
plt.title(f'Классификация k-NN (точность: {accuracy:.2f})')
plt.xlabel('X координата')
plt.ylabel('Y координата')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
