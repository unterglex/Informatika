```python
''' 1(5) ЗАДАНИЕ Вычислить скалярное произведение двух векторов размером 4. 
(Скалярным произведением называется сумма попарных произведений соответсвующих элементов массивов)'''

a = [4, 3, 6, 7]
b = [3, 3, 0, 3]
s = a[0]*b[0]+a[1]*b[1]+a[2]*b[2]+a[3]*b[3]
print(s)
exp_s = 4*3+3*3+6*0+7*3
if s == exp_s:
    print('Верно')
else: print('Неверно')
```

```python
''' 2(10) ЗАДАНИЕ Удалить минимальный среди положительных элементов массива. '''

m = [-3, -2, 5, 1, -4, 2, 2, -3, -1]
min_positive = float('inf')
for i in m:
    if i > 0 and i < min_positive:
        min_positive = i
if min_positive == float('inf'):
    print("В массиве нет положительных элементов")
else:
    m.remove(min_positive)
    print(f"Минимальный положительный элемент ({min_positive}) удален")
    print(f"Обновленный массив: {m}")
```

```python
''' 3(11) ЗАДАНИЕ Вычислить значения функции у = cosx + xsinx в n точках отрезка [a, b]. 
Результат получить в двух массивах Х (аргумент), У (функция). Используя сформированные массивы, определить все значения аргумента, 
при которых функция имеет максимум или минимум (включая локальные). 
Для точек, в которых достигается экстремум, вывести значения аргумента, функции и вид эстремума (максимум или минимум). 
Определить глобальные экстремумы.
Указание. Функция имеет локальный максимум со значением у{i} (из массива У) при значении аргумента х{i}
(из массива Х), если значения функции у{i-1} и y{i+1} в двух соседних точках меньше у{i}. Аналогично для минимума. '''

import math
n=int(input('Введите количество точек: '))
a=int(input('Введите левую границу отрезка [a,b]: '))
b=int(input('Введите правую границу отрезка [a,b]: '))
print() #Оформление вывода
print('x     y               min or max') #Оформление вывода
print() #Оформление вывода
h=(b-a)/n #Равномерный шаг на отрезке
X=[]
Y=[]
for i in range(n):
    X.append(a+i*h)
    Y.append(math.cos(X[i])+X[i]*math.sin(X[i]))
l_min=[] #Локальный минимум
l_max=[] #Локальный максимум
for i in range(1, len(Y)-1): #C мат. т. зрения крайние точки отрезка брать нельзя
    if Y[i]>Y[i-1] and Y[i]>Y[i+1]:
        print(X[i], Y[i], "Локальный максимум")
    elif Y[i]<Y[i-1] and Y[i]<Y[i+1]:
        print(X[i], Y[i], "Локальный минимум")
g_min = min(Y) #Глобальный минимум
min_x = []
for i in range(len(Y)):
    if Y[i]==g_min:
        min_x.append(X[i]) #Значение Х при глобальном минимуме
g_max = max(Y) #Глобальный максимум
max_x = []
for i in range(len(Y)):
    if Y[i]==g_max:
        max_x.append(X[i]) #Значение Х при глобальном максимуме
print() #Оформление вывода
print("Глобальный минимум: ", g_min, 'достигается при x =', min_x)
print("Глобальный максимум: ", g_max, 'достигается при x =', max_x)
```
