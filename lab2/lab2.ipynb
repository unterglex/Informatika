```python
'''1(5) Заданы площади: круга - r и квадрата - s. Определить, поместится ли квадрат в круге. 
Задачу решить при 1)r=70; s=36,74; 2)r=0,86; s=0,74.'''
r=int(input('Введите площадь круга:'))
s=int(input('Введите площадь квадрата:'))
import math
Rr = math.sqrt(r/math.pi)
Sr = math.sqrt(s)/2
if Rr > Sr:
    print("Квадрат поместится в круг")
else: print("Квадрат не сможет поместиться в круг")
r=70
s=36.74
Rr = math.sqrt(r/math.pi)
Sr = math.sqrt(s)/2
if Rr > Sr:
    print("Квадрат c площадью 36.74 поместится в круг площадью 70")
else: print("Квадрат с площадью 36.74 не сможет поместиться в круг площадью 70")
r=0.86
s=0.74
Rr = math.sqrt(r/math.pi)
Sr = math.sqrt(s)/2
if Rr > Sr:
    print("Квадрат c площадью 0.74 поместится в круг площадью 0.86")
else: print("Квадрат с площадью 0.74 не сможет поместиться в круг площадью 0.86")
```

```python
'''2(5)В соревнованиях по бегу принимают участие 30 спортсменов. 
Вводя по очереди результаты участников, определить, сколько из них выполнили заданный норматив.'''
x=0
nado=int(input('Введите время норматива в секундах: '))
for i in range(30):
    rez=int(input('Введите результат спортсмена: '))
    if rez<=nado:
        x+=1
print('Норматив выполнили', x, 'человек')
```

```python
''' 2(11) В группе учится n студентов. Каждый сдал 4 экзамена. Подсчитать число неуспевающих студентов и средний балл группы. Количество данных заранее неизвестно. 
В программе необходимо обеспечить прекращение ввода, как только входной поток иссякнет '''
ex1=int(input('Введите минимальный балл по первому экзамену: '))
ex2=int(input('Введите минимальный балл по второму экзамену: '))
ex3=int(input('Введите минимальный балл по третьему экзамену: '))
ex4=int(input('Введите минимальный балл по четвертому экзамену: '))
x=0
s=0
kolvo=0
a=True
while a:
    name = input('Введите имя ученика (поставьте 0, чтобы завершить программу):')
    if name == '0':
        break
    res1=int(input('Введите результат первого экзамена: '))
    res2=int(input('Введите результат второго экзамена: '))
    res3=int(input('Введите результат третьего экзамена: '))
    res4=int(input('Введите результат четвертого экзамена: '))
    if res1<=ex1 or res2<=ex2 or res3<=ex3 or res4<=ex4:
        x+=1
    s+=res1+res2+res3+res4
    kolvo+=1
sb=s/(kolvo*4)
print('Количество неуспевающих студентов: ', x)
print('Средний балл группы:', sb)
```
